```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
%matplotlib inline
plt.rcParams['text.usetex'] = True
```

```python
def f(x):
    return x * math.sqrt(abs(x)) / (2 + math.sin(x))
```

```python
xs = np.arange(-5, 5, 1)
f = np.vectorize(f)
ys = f(xs)
```

```python
plt.plot(xs, ys)
```

```python
fig, ax = plt.subplots(figsize=(10, 7))
# Axes
ax.plot(xs, ys, label='f(x)', color='red', marker='v')
# ax.set_title(r'Some weird function $f(x) = \frac{x \sqrt{x}}{2 + \sin x}$')
ax.set_title(r'Some weird function')
ax.grid(axis='both')

ax.legend(fontsize=16, loc='lower right', labels=['f(x)'], title='Legend')
```

```python
theta = np.polyfit(xs, ys, deg=10)
model = np.poly1d(theta)
xnew = np.arange(-5, 5, 0.01)
plt.plot(xnew, model(xnew))
plt.plot(xnew, f(xnew), color='red')
```
